##### LEAST SQUARES FIT, Relationship between variables (unemployment and weekly sales)

```python
inter, slope = thinkstats2.LeastSquares(unemp, wklysls)
inter, slope
```

With a slope **-289**, it means the weekly sales decline by **-$289.34** for every increase in unemployment

```python
inter + slope * 22
```

the higher the unemployment rate the lower the weekly sales

```python
fit_xs, fit_ys = thinkstats2.FitLine(unemp, inter, slope)
```

```python
plt.figure(figsize=(14,7))
thinkplot.Scatter(unemp, wklysls, color='blue', alpha=0.1, s=10)
thinkplot.Plot(fit_xs, fit_ys, color='white', linewidth=3)
thinkplot.Plot(fit_xs, fit_ys, color='red', linewidth=2)
thinkplot.Config(xlabel="Unemployment Rate",
                 ylabel='Weekly Sales',
                 legend=True)
```

From the plot we can see that the slope line decreases with value increases in employment rate

```python
sample['residual'] = thinkstats2.Residuals(sample['Unemployment'], sample['Weekly_Sales'], inter, slope)
```

```python
bins = np.arange(1, 15, 2)
indices = np.digitize(sample.Unemployment, bins)
groups = sample.groupby(indices)

unemp_means = [group.Unemployment.mean() for _, group in groups][1:-1]
unemp_means
```

```python
cdfs = [thinkstats2.Cdf(group.residual) for _, group in groups][1:-1]
```

```python
def PlotPercentiles(unemp_means, cdfs):
    thinkplot.PrePlot(3)
    for percent in [75, 50, 25]:
        sales_percentiles = [cdf.Percentile(percent) for cdf in cdfs]
        label = '%dth' % percent
        thinkplot.Plot(unemp_means, sales_percentiles, label=label)
```

```python
PlotPercentiles(unemp_means, cdfs)

thinkplot.Config(xlabel="Unemployment Rate",
                 ylabel='Residuals ($)',
                 xlim=[1, 15])
```

##### Let run a Hypothesis test to test the difference in weekly sales mean for holiday and non-holiday periods

```python
class DiffWeeklySalesMeansPermute(thinkstats2.HypothesisTest):
    """Tests a difference in means by permutation."""

    def TestStatistic(self, data):
        """Computes the test statistic.
        data: data in whatever form is relevant        
        """
        group1, group2 = data
        test_stat = abs(group1.mean() - group2.mean())
        return test_stat

    def MakeModel(self):
        """Build a model of the null hypothesis.
        """
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        """Run the model of the null hypothesis.
        returns: simulated data
        """
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
    
    
class DiffWeeklySalesStdPermute(thinkstats2.HypothesisTest):
    """Tests a difference in means by permutation."""

    def TestStatistic(self, data):
        """Computes the test statistic.
        data: data in whatever form is relevant        
        """
        group1, group2 = data
        test_stat = abs(group1.std() - group2.std())
        return test_stat

    def MakeModel(self):
        """Build a model of the null hypothesis.
        """
        group1, group2 = self.data
        self.n, self.m = len(group1), len(group2)
        self.pool = np.hstack((group1, group2))

    def RunModel(self):
        """Run the model of the null hypothesis.
        returns: simulated data
        """
        np.random.shuffle(self.pool)
        data = self.pool[:self.n], self.pool[self.n:]
        return data
    
    

class CorrelationPermute(thinkstats2.HypothesisTest):

    def TestStatistic(self, data):
        xs, ys = data
        test_stat = abs(thinkstats2.Corr(xs, ys))
        return test_stat

    def RunModel(self):
        xs, ys = self.data
        xs = np.random.permutation(xs)
        return xs, ys
    
```

```python
holiday_sales = holiday_groups[True]['Weekly_Sales']
non_holiday_sales = holiday_groups[False]['Weekly_Sales']
holiday_data = holiday_sales, non_holiday_sales
```

```python
ht = DiffWeeklySalesMeansPermute(holiday_data)
```

```python
ht.PValue()
```

Here the p value is 0 meaning that there is some significance to the assumption that weekly sales increase during the holiday periods

```python
ht.PlotCdf()
thinkplot.Config(xlabel='test statistic',
                   ylabel='CDF')
```

Above is the CDF showing the distribution of weekly sales means over the many simulated samples

```python
print('Actual:', ht.actual, 'Largest from simulation',  ht.MaxTestStat())
```

Lets run a simlar test again but this time we will use the standard deviation rather than the means of the weekly sales

```python
ht2 = DiffWeeklySalesStdPermute(holiday_data)
```

```python
ht2.PValue()
```

```python
ht2.PlotCdf()
```

```python
print('Actual:', ht2.actual, 'Largest from simulation',  ht2.MaxTestStat())
```
